```python
# Load the dataset
learning_share = pd.read_csv('District_Monthly_Shares_03.08.23.csv')

# Display the first few rows and columns to understand the structure
learning_share.head()
```

```python
# Load the dataset

seda2022 = pd.read_csv('seda2022_admindist_poolsub_ys_beta.csv')

# Display the first few rows and columns to understand the structure
seda2022.head()
```

```python
# Set treatment variable D as an indicator: threshhold = 0.5
learning_share["D"] = ((learning_share["share_virtual"] + learning_share["share_hybrid"]) >= 0.5).astype(int)

# Sum up the treatment variable D
learning_share_df = learning_share.groupby(["StateAbbrev", "NCESDistrictID", "DistrictName"], as_index=False)["D"].sum()

# Arrange columns
# Drop the columns: month, share_inperson, share_hybrit, share_virtual
learning_share_df = (learning_share_df.rename(columns={"D": "Online_months"})
                     .sort_values('NCESDistrictID', ascending=True)).reset_index(drop=True)

# 結果をCSVとして保存
learning_share_df.to_csv("online_share_months.csv", index=False)

learning_share_df.head(10)
```

```python
# Filter for rows where subgroup == "all" and drop 'fips' and 'subgroup' columns
seda_filtered = (seda2022[seda2022["subgroup"] == "all"].drop(columns=["fips", "subgroup"])
                 .sort_values('sedaadmin', ascending=True)).reset_index(drop=True)

# Save the filtered DataFrame to CSV
seda_filtered.to_csv("seda_2022.csv", index=False)

seda_filtered.head(10)
```

```python
ls_share = pd.read_csv("online_share_months.csv")
ls_share["DistrictName"] = ls_share["DistrictName"].str.upper()

seda_df = pd.read_csv('seda_2022.csv')
seda_df["sedaadminname"] = seda_df["sedaadminname"].str.upper()

display(ls_share.head())
display(seda_df.head())
```

```python
# Combin the two dataframe with (District ID and District Name)
common_states = list(set(ls_share['StateAbbrev'].unique()) & set(seda_df['stateabb'].unique()))

ls_share = ls_share[ls_share['StateAbbrev'].isin(common_states)]
seda_df = seda_df[seda_df['stateabb'].isin(common_states)]

merged_df = pd.merge(
    ls_share,
    seda_df,
    how="inner",
    left_on=["NCESDistrictID", "DistrictName"],
    right_on=["sedaadmin", "sedaadminname"]
)

merged_df = merged_df.drop(columns=['StateAbbrev', 'NCESDistrictID', 'DistrictName']).dropna(axis=1).reset_index(drop=True)
print(len(merged_df))
merged_df.head()
```

```python
math_df = merged_df[merged_df['subject'] == 'mth'].drop(columns=['subject']).reset_index(drop=True)
display(math_df.head())
math_df.to_csv('math_seda.csv', index=False)

rla_df = merged_df[merged_df['subject'] == 'rla'].drop(columns=['subject']).reset_index(drop=True)
display(rla_df.head())
rla_df.to_csv('rla_seda.csv', index=False)
```
